## Graph analysis in Python with [NetworkX](https://networkx.org/)


NetworkX is good for small computations, but for big data networks it's much less efficient and it takes a lot of time. More famous network analysis tools can be done with [iGraph](https://igraph.org/). Also one library done by the lecturer - [CNetwork](https://github.com/VictorSeven/CNetwork).

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline
```

```python
network = nx.Graph() # abstract representation of a graph
```

<img src="graph_types.png"
     alt="Markdown Monster icon"
     style="float: left; margin-right: 10px;" />

```python
network.add_nodes_from([("ada", {"surname": "altieri"}), ("bob", {"surname": "sack"})]) #list of nodes with properties
network.add_edges_from([("ada", "bob"), ("ada", "clarck")]) #list of tuples as edges

# remove the names from nodes and convert them into ordered list of indices
network = nx.convert_node_labels_to_integers(network, label_attribute="old_name")
# equivalent to network.add_nodes_from((0, {"old_name": "ada"})...)

print(network.number_of_nodes(), "nodes and", network.number_of_edges(), "edges in a network")
```

```python
print(network.nodes()[0]) # returns a NodeView 
print(network.nodes(data=True)) # view on all characteristic of a graph
```

```python
# iterating over graphs
for index, attrib in network.nodes(data=True):
    print("Index {i} is called {name}".format(i=index, name=attrib["old_name"]))
```

```python
n = 1000 # number of nodes
k = 10 # degree
p = k / n # probability 

erdos_renyi = nx.Graph() # initialize
erdos_renyi.add_nodes_from(range(n)) # network with n nodes ordered from 0 to 999

# create connections randomly with a probability p
for i in range(n): 
    for j in range(i+1, n):
        if np.random.rand() < p:
            erdos_renyi.add_edge(i, j)
```

```python
degree_distribution = nx.degree_histogram(erdos_renyi) 
degrees = range(len(degree_distribution))

plt.figure()
plt.bar(degrees, degree_distribution)
plt.xlabel("k")
plt.ylabel("P(k)")
plt.show()

# it's Poisson distribution, only similar to Gaussian
```

Gaussian distribution is characterized by the possibility of having different mean and variance, while Poisson distribution has the same mean and the variance (one of the most important conceptual difference). 

```python
# we expect very low, around 0, because it's a random network

print("clustering =", nx.average_clustering(erdos_renyi))
```
